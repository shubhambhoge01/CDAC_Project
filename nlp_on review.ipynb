```python
import opendatasets as od
import pandas as pd
import nltk
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
od.download("https://www.kaggle.com/yelpprojectpurpose/yelp-dataset-for-project")
```

```python
review = pd.read_csv("./yelp-dataset-for-project/Review.csv")
```

```python
vader=SentimentIntensityAnalyzer()
response = []
```

```python
for i in review['text']:
    feedback = vader.polarity_scores(i)
    if feedback['compound']<0 :
        ans = 'Negative'
    elif feedback['compound']>0:
        ans = 'Positive'
    response.append(ans)    
```

```python
len(response)
```

```python
res = pd.DataFrame({'response':response})
```

```python
review['response']= res.response
```
